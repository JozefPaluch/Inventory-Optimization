```python
import pandas as pd
import numpy as np
from faker import Faker
import random
from datetime import datetime, timedelta
```

```python
# wymyślone dane
fake = Faker()
Faker.seed(42)
np.random.seed(42)
```

```python
num_products = 100
categories = ['Electronics', 'Furniture', 'Clothing', 'Home & Garden', 'Toys']
```

```python
products_data = []

for i in range(1, num_products + 1):
    product_id = i
    category = random.choice(categories)
    
    # cena mniej więcej dobrana do kategorii
    if category == 'Electronics':
        price = round(np.random.uniform(100, 2000), 2)
    elif category == 'Furniture':
        price = round(np.random.uniform(50, 800), 2)
    else:
        price = round(np.random.uniform(10, 150), 2)
        
    products_data.append({
        'product_id': product_id,
        'product_name': f"{category} - {fake.word().capitalize()} {fake.color_name()}",
        'category': category,
        'unit_price': price,
        'safety_stock': random.randint(5, 20)
    })
```

```python
# wrzucenie do pliku csv
df_products = pd.DataFrame(products_data)
df_products.to_csv('products.csv', index=False)
```

```python
# rok działania wymyślonego magazynu
start_date = datetime(2023, 1, 1)
days_to_simulate = 365
movements_data = []
```

```python
for pid in df_products['product_id']:
    movements_data.append({
        'date': start_date,
        'product_id': pid,
        'type': 'PZ', # dostaw
        'quantity': random.randint(20, 100)
    })
```

```python
current_date = start_date
for day in range(days_to_simulate):
    current_date += timedelta(days=1)
    daily_transactions = random.randint(10, 50)
    
    for _ in range(daily_transactions):
        prod_id = random.randint(1, num_products)
        
        # 80% szans na sprzedaz, 20% szans na dostawę
        if random.random() > 0.2:
            move_type = 'WZ'
            qty = random.randint(1, 5) * -1 
        else:
            move_type = 'PZ'
            qty = random.randint(10, 30)
            
        movements_data.append({
            'date': current_date,
            'product_id': prod_id,
            'type': move_type,
            'quantity': qty
        })
```

```python
# plik operacji magazynowych
df_movements = pd.DataFrame(movements_data)
df_movements.to_csv('movements.csv', index=False)
```
